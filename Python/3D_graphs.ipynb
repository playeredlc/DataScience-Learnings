```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D #https://matplotlib.org/mpl_toolkits/mplot3d/
from matplotlib import cm # color map
import numpy as np
```

---
#### 3D graphs
---
##### *Minimise:*

##### $f(x, y) = \frac{1}{3^{x² - y²} + 1}$

```python
def f(x, y):
    pol = pow(3, (-x**2 - y**2))
    return 1 / (pol + 1)
```

```python
# generate data
x_1 = np.linspace(start=-2, stop=2, num=200)
y_1 = np.linspace(start=-2, stop=2, num=200)
print(x_1.shape)
# x_1 and y_1 are 1d numpy arrays. (200 rows, 0 col)
# plot_surface method expects 2d arrays as arguments:
x_1, y_1 = np.meshgrid(x_1, y_1)
print(x_1.shape)
```

---
##### Generating a 3D plot:

```python
plt.style.use('dark_background')

fig = plt.figure(figsize=[16, 12]) #figure object (matplotlib.figure.Figure)
ax = fig.gca(projection='3d') # get current axes (create if none)

ax.set_xlabel('X', fontsize=16)
ax.set_ylabel('Y', fontsize=16)
ax.set_zlabel('f(x,y) = cost', fontsize=16)
ax.plot_surface(x_1, y_1, f(x_1, y_1), cmap=cm.coolwarm, alpha=0.5)
```
