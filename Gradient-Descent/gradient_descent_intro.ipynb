```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import random
```

---

#### Cost function:
##### $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

##### Generate some data and visualize:
    numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
##### *'Returns num evenly spaced samples, calculated over the interval [start, stop].'*

```python
x_1 = np.linspace(-3, 3, num=500)
```

```python
plt.style.use('dark_background')
plt.figure(figsize=[10, 5])
plt.xlim(-3, 3)
plt.ylim(0, 13)
plt.title('Cost function', fontsize=16, pad=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.plot(x_1, f(x_1), c='b')
```

---

##### *Derivatives and slope:*
##### $ \frac{d}{dx}f(x) = 2x + 1$

```python
def df(x):
    return 2*x + 1
```

```python
plt.style.use('dark_background')
plt.figure(figsize=[15, 5])

# cost function graph
cost = plt.subplot(1, 2, 1) #1row, 2cols, index=1
plt.title('Cost function', fontsize=16, pad=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.xlim(-3, 3)
plt.ylim(0, 14)
plt.grid()
cost.plot(x_1, f(x_1), c='b')

# derivative graph
deriv = plt.subplot(1, 2, 2) #1row, 2cols, index=2
plt.title('Derivative of cost function', fontsize=16, pad=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('df(x)', fontsize=12)
plt.xlim(-3, 3)
plt.ylim(-6, 8)
plt.grid()
deriv.plot(x_1, df(x_1), c='r')

plt.show()
```

---

#### Gradient descent algorithm:

```python
# algorithm variables
new_x = random.uniform(-100, 100)
learning_rate = 0.1
n_steps = 100
precision = 0.0001

# visualization variables
x_list = [new_x]
slope_list = [df(new_x)]
```

```python
for i in range(n_steps):
    prev_x = new_x # hold the value of previous iteration
    gradient = df(prev_x) # update the slope value
    new_x = prev_x - (learning_rate * gradient) # go one step further towards the minimun
    x_variation = abs(new_x - prev_x) # check progress (step size)
    
    #store data for visualization
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    # stop criteria
    if x_variation < precision:
        print('Loop amount: ', i)
        break
print('Local min: ', new_x)
print('Slope at this point: ', df(new_x))
print('Cost at this point: ', f(new_x))
```

---

##### *Visualization of the data generated from the algorithm:*

```python
plt.style.use('dark_background')
plt.figure(figsize=[15, 10])

# cost function graph
cost = plt.subplot(2, 2, 1) #2rows, 2cols, index=1
plt.title('Cost function', fontsize=16, pad=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.xlim(-3, 3)
plt.ylim(0, 14)
plt.grid()
x_array = np.array(x_list)
plt.scatter(x_array, f(x_array), c='r', s=80) # INSERT DATA FROM GRADIENT DESCENT
cost.plot(x_1, f(x_1), c='b')

# derivative graph
deriv = plt.subplot(2, 2, 2) #2rows, 2cols, index=2
plt.title('Derivative of cost function', fontsize=16, pad=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('df(x)', fontsize=12)
plt.xlim(-3, 3)
plt.ylim(-6, 8)
plt.grid()
plt.scatter(x_list, slope_list, c='b', s=80) # INSERT DATA FROM GRADIENT DESCENT
deriv.plot(x_1, df(x_1), c='r')

# cost function graph (close up)
cost_closeup = plt.subplot(2, 2, 3) #2rows, 2cols, index=3
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.xlim(-0.6, -0.4)
plt.ylim(0.65, 0.85)
plt.grid()
x_array = np.array(x_list)
plt.scatter(x_array, f(x_array), c='r', s=80) # INSERT DATA FROM GRADIENT DESCENT
cost_closeup.plot(x_1, f(x_1), c='b')

# derivative graph (closeup)
deriv_closeup = plt.subplot(2, 2, 4) #2rows, 2cols, index=2
plt.xlabel('x', fontsize=12)
plt.ylabel('df(x)', fontsize=12)
plt.xlim(-0.6, -0.4)
plt.ylim(-0.10, 0.10)
plt.grid()
plt.scatter(x_list, slope_list, c='b', s=80) # INSERT DATA FROM GRADIENT DESCENT
deriv_closeup.plot(x_1, df(x_1), c='r')



plt.show()
```

##### __*Cost graph:*__ Shows the f(x) value getting iteratively closer to the minimun.
##### __*Slope graph:*__ Shows the slope getting iteratively closer to 0. Which means f(x) is getting closer to the minimun.
