```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

---

##### *Store cleaned data inside 'data'*

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

##### *Some info about the data loaded:*

```python
data.describe()
```

---

##### *X is the feature (the production budget); y is the target(worldwide gross).*

```python
X = DataFrame(data, columns=['production_budget_usd'])
```

```python
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

---

##### *Ploting the data for better visualization*

```python
# Sizing
plt.figure(figsize=(10,6))
plt.ylim(0, 3e09)
plt.xlim(0, 4.5e08)
# Labeling
plt.title('Movies Cost vs Global Revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide Gross $')
# Plotting
plt.scatter(X, y, alpha=0.3)
plt.show()
```

---

##### *LinearRegression fits a linear model to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation.*

```python
linearRegression = LinearRegression()
linearRegression.fit(X, y)
```

##### *Slope coefficient: Estimated coefficients for the linear regression problem.*

```python
linearRegression.coef_ # theta_01
```

##### *Independent term in the linear model.*

```python
linearRegression.intercept_ #theta_02
```

##### *(Score, R², Goodness of fit) coefficient of determination of the prediction.*
##### *The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).*

```python
# it means the feature(production budget) can explain around 55% of the variation of the revenue
linearRegression.score(X, y)
```

---

##### *Draw the line using the model prediction on the graph above*

```python
# Sizing
plt.figure(figsize=(10,6))
plt.ylim(0, 3e09)
plt.xlim(0, 4.5e08)
# Labeling
plt.title('Movies Cost vs Global Revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide Gross $')
# Plotting
plt.scatter(X, y, alpha=0.3)

# Draw line
plt.plot(X, linearRegression.predict(X), color='red', linewidth=2) # draw line (y value [2nd param] is predicted by the algorithm)

plt.show()
```
